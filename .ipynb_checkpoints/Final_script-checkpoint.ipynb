```python
#This is the final script that recommends recipes and shopping lists based on user nutritional requirements
#AND includes a recommendation engine to match user preferences

#FIXME: 
#1) Thiss makes a lot of API queries. May be possible to cache data, or ask Spoonacular for custom endpoints
#3) The bf/lunch/dinner breakdwon needs to be corrected with appropriate values
#4) Speak to nutritionist to add more user input (e.g: BMI, whether they want to lose weight, breastfeeding...)
#5) Recipes in Spooncacular are in American units (cups, etc) At the moment we are showing ingredients in the metric system, but 
#some ingredients expressed in cups (unit of volume) in the US (e.g. tofu), should be expressed in g (unit of mass) in the UK. 
#This results in "mL of tofu" of the shopping list. Unsure whether we should express in cups instead
#6) the recommendation engine is based on the cosine distance between user vector ans recipe features. 
#We need to define an appropriate threshold for this, maybe by asking user how open they are to new foods


import numpy as np
import pandas as pd
import os
import requests
import json
import scipy.spatial.distance



#defines function to convert user input to boolean
def str2bool(v):
  return v.lower() in ("yes", "true", "t", "1","y")
```

```python
#this function returns the feature matrix for the json object returned by the API query

def create_features_matrix(Train_Recipes_Info):
    #adds the tags columns
    features=pd.DataFrame([])
    features['id'] = pd.Series(0, index=features.index)
    features['cheap'] = pd.Series(0, index=features.index)
    features['veryPopular'] = pd.Series(0, index=features.index)
    features['lowFodmap'] = pd.Series(0, index=features.index)
    features['glutenFree'] = pd.Series(0, index=features.index)
    features['vegan'] = pd.Series(0, index=features.index)
    features['veryHealthy'] = pd.Series(0, index=features.index)
    features['weightWatcherSmartPoints'] = pd.Series(0, index=features.index)
    features['aggregateLikes'] = pd.Series(0, index=features.index)
    features['vegetarian'] = pd.Series(0, index=features.index)
    features['ketogenic'] = pd.Series(0, index=features.index)
    features['readyInMinutes'] = pd.Series(0, index=features.index)
    features['servings'] = pd.Series(0, index=features.index)
    features['sustainable'] = pd.Series(0, index=features.index)
    features['dairyFree'] = pd.Series(0, index=features.index)
    features['whole30'] = pd.Series(0, index=features.index)
    #adds the ingredient aisle columns
    for j in Train_Recipes_Info:
        for i in j['extendedIngredients']:
            #sometimes the ingredient doesn't have an aisle
            try:
                #if it does, add it to the df
                if not i['aisle'] in features.columns:
                    features[i['aisle']] = pd.Series(0, index=features.index)
            except:
                print "WARNING: the aisle is missing!"
        #adds all values to the df
        tmp=pd.DataFrame([np.zeros(len(features.columns.values)).tolist()],columns=features.columns)
        features=features.append(tmp,ignore_index=True)
        features.iloc[-1,features.columns.get_loc('id')]=j['id']
        features.iloc[-1,features.columns.get_loc('cheap')]=j['cheap']
        features.iloc[-1,features.columns.get_loc('veryPopular')]=j['veryPopular']
        features.iloc[-1,features.columns.get_loc('lowFodmap')]=j['lowFodmap']
        features.iloc[-1,features.columns.get_loc('glutenFree')]=j['glutenFree']
        features.iloc[-1,features.columns.get_loc('vegan')]=j['vegan']
        features.iloc[-1,features.columns.get_loc('veryHealthy')]=j['veryHealthy']
        features.iloc[-1,features.columns.get_loc('weightWatcherSmartPoints')]=j['weightWatcherSmartPoints']
        features.iloc[-1,features.columns.get_loc('aggregateLikes')]=j['aggregateLikes']
        features.iloc[-1,features.columns.get_loc('vegetarian')]=j['vegetarian']
        features.iloc[-1,features.columns.get_loc('ketogenic')]=j['ketogenic']
        features.iloc[-1,features.columns.get_loc('readyInMinutes')]=j['readyInMinutes']
        features.iloc[-1,features.columns.get_loc('servings')]=j['servings']
        features.iloc[-1,features.columns.get_loc('sustainable')]=j['sustainable']
        features.iloc[-1,features.columns.get_loc('dairyFree')]=j['dairyFree']
        features.iloc[-1,features.columns.get_loc('whole30')]=j['whole30']
        
        for i in j['extendedIngredients']:
            try:
                features.iloc[-1,features.columns.get_loc(i['aisle'])]+=1   
            except:
                #sometimes the ingredient does not have an aisle :-(
                pass
            
    features=features.set_index('id')
    return features
```

```python
#master lists of dataframes for lunch and dinner. This is so we can switch between the lunch and dinner DFs
Lunch_Dinner_Df_List=[["lunch"], ["dinner"]]

#This checks if the user info files already exist
if os.path.isfile('user_info.dat'):
    #if file exists then load files
    #read usr info
    UserAge,UserActivity=np.loadtxt('user_info.dat',skiprows=1,unpack=True,dtype=int)  
    #read utility matrix
    Utility_Df_Lunch=pd.read_csv('lunch_utility_Df.csv').set_index('Id')
    Utility_Df_Dinner=pd.read_csv('dinner_utility_Df.csv').set_index('Id')
    #read features matrix
    features_Df_Lunch=pd.read_csv('lunch_features.csv').set_index('id')
    features_Df_Dinner=pd.read_csv('dinner_features.csv').set_index('id')
    #adds them to the master lists
    Lunch_Dinner_Df_List[0].append(Utility_Df_Lunch)
    Lunch_Dinner_Df_List[0].append(features_Df_Lunch)
    Lunch_Dinner_Df_List[1].append(Utility_Df_Dinner)   
    Lunch_Dinner_Df_List[1].append(features_Df_Dinner)
else:
    #if the files dont exist, we create them!
    UserAge = raw_input("How old are you?")
    UserActivity= raw_input("How active are you on a scale of 1 to 3?")
    
    #save usr info
    np.savetxt('user_info.dat',(["age","activity"],[str(UserAge),str(UserActivity)]), delimiter=" ", fmt="%s")
    
    #load recipes in 'training' set
    for matrices in Lunch_Dinner_Df_List:
        #load the training recipies to train the recomendation engine on
        Train_Recipes_Info=json.loads(open(matrices[0]+'_good_recipes.json', "r").read())
        #build features matrix
        features_df=create_features_matrix(Train_Recipes_Info)
        #build utility matrix
        n=-1
        Recipe_Ids=[]
        #print the training set
        for j in Train_Recipes_Info:
            n+=1
            Recipe_Ids.append(j['id'])
            print "#", n,  ":",  j['title'] , "(", j['spoonacularSourceUrl'],")"
        #ask which ones they like.
        Liked = raw_input("Please enter the id of the recipes that you like for "+ str(matrices[0])+", separated by commas")
        #add each recipe to utility matrix
        Utility_Matrix = -np.ones(len(Recipe_Ids))
        for i in range(0, len(Recipe_Ids)) :
            if str(i) in Liked:
                Utility_Matrix[i]+=2
        Recipe_Ids=np.array(Recipe_Ids)      
        Utility_Df=pd.DataFrame(np.concatenate(([Recipe_Ids], [Utility_Matrix])).T, columns=["Id", "Like"])
        Utility_Df=Utility_Df.set_index('Id')
        #add matricies to master list
        matrices.append(Utility_Df)
        matrices.append(features_df)
        #save utility matrix and usr vector
        
        features_df.to_csv( str(matrices[0])+'_features.csv')
        Utility_Df.to_csv( str(matrices[0])+'_utility_Df.csv')
```

```python
#based on user info, calculate cal requirements
Cal_Table=np.loadtxt('table2.dat')
Cal_Frame=pd.DataFrame(Cal_Table,columns=['age','m1','m2','m3','f1','f2','f3'])
User_Activity_Sex='f'+str(UserActivity)#FIXME: right now it assumrs user is female

Tot_Cal= Cal_Frame[Cal_Frame.age<=float(UserAge)][User_Activity_Sex].values[-1]

Breakdown = np.array([0.25,0.37,0.37])#FIX ME: add correct values for meal breakdown
Cal_Per_Meal=Breakdown * Tot_Cal

fatsFrac=np.array([0.20,0.35])
carbFrac=np.array([0.40,0.65])
protFrac=np.array([0.1,0.35])

Lunch_cals= Cal_Per_Meal[1]
Lunch_carbs=Cal_Per_Meal[1]*carbFrac/4. #4cals/g in carbs
Lunch_prot=Cal_Per_Meal[1]*protFrac/4. #4cals/g in prots
Lunch_fats=Cal_Per_Meal[1]*protFrac/9.  #9cals/g in fats

Dinner_cals= Cal_Per_Meal[2]
Dinner_carbs=Cal_Per_Meal[2]*carbFrac/4. #4cals/g in carbs
Dinner_prot=Cal_Per_Meal[2]*protFrac/4. #4cals/g in prots
Dinner_fats=Cal_Per_Meal[2]*protFrac/9.  #9cals/g in fats

#add the requirments to the master list

Lunch_Dinner_Df_List[0].append([Lunch_cals, Lunch_carbs, Lunch_fats, Lunch_prot])
Lunch_Dinner_Df_List[1].append([Dinner_cals, Dinner_carbs,  Dinner_fats, Dinner_prot])
```

```python
#this loop runs API queries that match nutritional requirements and computes cosine distance between recipe and user vector
#it keeps recipes with low distance (high simularity), suggests to user, and offers option to keep or reject
# it updating the utility matrix and features matrix according to user preference

#the number of days that recipes are required
Days=int(raw_input("Please enter the number of days required for the shopping list (max=7 days)"))
all_good_recipes=[] #this will contain alll the recipes to store for the shopping list for lunch AND dinner
for meal in Lunch_Dinner_Df_List: #This for loop is so we can first get recipes for lunch and then for dinner
    n=meal[3] #this is the nutrisional requirments for the user for that meal
    Utility_Df=meal[1]
    features_df=meal[2]
    Good_Recipes=[] #these are the recipes we will store for the ingredients list for lunch, then for dinner
    while len(Good_Recipes) < int(Days):
        query_request="https://spoonacular-recipe-food-nutrition-v1.p.mashape.com/recipes/findByNutrients?maxCalories="+str(int(n[0]+50.))+"&maxCarbs="+str(int(n[1][1]))+"&maxFat="+str(int(n[2][1]))+"&maxProtein="+str(int(n[3][1]))+"&minCalories="+str(int(n[0]-50.))+"&minCarbs="+str(int(n[1][0]))+"&minFat="+str(int(n[2][0]))+"&minProtein="+str(int(n[3][0]))
        response = requests.get(query_request, headers={"X-Mashape-Key": }).json() #ADD SPOONACULAR LOGIN KEY
        Recipe_Ids=[]
        for i in response:
            Recipe_Ids.append(i['id'])
        Recipes_Info=[]
        for j in Recipe_Ids:
            Means= features_df.mean() #the mean of each aisle in the features matrix
            Sds= features_df.std() #the standard dev of each aisle in the features matrix
            Sds[Sds==0]=1 #set the std to 1 if all tags in one column in features matrix are same.
            Z_scores = features_df.subtract(Means,axis=1).divide(Sds,axis=1) #re-normalise the features matrix
            Merged= Utility_Df.merge(Z_scores,right_index=True,left_index=True).fillna(0) #combine features and utility df
            usr_vector=Merged.multiply(Merged["Like"], axis="index").mean() #make usr vector
            usr_vector=pd.DataFrame(usr_vector.drop("Like")).transpose()
            
            #query API to get new recipe full details
            Query_Request_Info="https://spoonacular-recipe-food-nutrition-v1.p.mashape.com/recipes/"+str(j)+"/information?includeNutrition=false"
            Recipe_Response = requests.get(url=Query_Request_Info, headers={"X-Mashape-Key": }).json()#ADD SPOONACULAR LOGIN KEY
            #Create feature matrix for new recipe
            Recipe_Features=create_features_matrix([Recipe_Response])
            #normalising feature matrix for new recipe using mean and sd from old feature matrix
            tmp_Recipe_Features=Recipe_Features.append(pd.DataFrame(Means,columns=['mean']).transpose())
            tmp_Recipe_Features=tmp_Recipe_Features.append(pd.DataFrame(Sds,columns=['std']).transpose())
            tmp_Recipe_Features=tmp_Recipe_Features.transpose()
            nom_Recipe_Features=tmp_Recipe_Features.subtract(tmp_Recipe_Features['mean'],axis=0).divide(tmp_Recipe_Features['std'],axis=0)[j]          
            nom_Recipe_Features=pd.DataFrame(nom_Recipe_Features.fillna(0)).transpose()
            tmp_merge=usr_vector.append(nom_Recipe_Features).fillna(0)#append user vector and new recipe features
            
            Distance=scipy.spatial.distance.cosine(tmp_merge.values[0],tmp_merge.values[1]) #calculate the cosine distance between the usr vector and the new recipe. 0=same; 2=anti-parellel
            print Distance
            if Distance<1.2: #FIXME: how to choose appropriate distance?
                print Recipe_Response['title'],Recipe_Response['spoonacularSourceUrl']
                l= str2bool(raw_input('Would you like this recipe for '+meal[0]+"(y/n) ?")) # this is the function defined at the beginning of the script
                #ask usr if they are happy with recipe
                if l:
                    #if the ust likes it, add it to the recipies to keep
                    Good_Recipes.append(Recipe_Response)
                #update features matrix and utility matrix 
                features_df=features_df.append(Recipe_Features).fillna(0)
                tmp=pd.DataFrame([[Recipe_Response['id'],2*int(l)-1]],columns=["Id","Like"]).set_index('Id')
                Utility_Df=Utility_Df.append(tmp)
                if len(Good_Recipes)>=int(Days):
                    break
    #add selected recpies to a new recipies list
    all_good_recipes+=Good_Recipes
    #save utility matrix and feature matrix    
    meal[1]=Utility_Df
    meal[2]=features_df
    features_df=features_df *1
    features_df.to_csv( str(meal[0])+'_features.csv')
    Utility_Df.to_csv( str(meal[0])+'_utility_Df.csv')
```

```python
#appends ingredients to a PD DF

ingredients=[]
ingredients_to_check=[]

#we are now printing unitLong for clarity, but saving unitShort to file (more consistent than unit)
for i in all_good_recipes:  
    for j in i['extendedIngredients']:
        try:
            ingredients.append([j['amount'], j['unitShort'] ,j['name'], j['id']])
            #print j['amount'], j['unitLong'] ,j['name'], j['id']
        except:
            print "could not find  ", j['name']
            ingredients_to_check.append(j['name'])
            
ingredients_df=pd.DataFrame(ingredients,columns=['amount','unit','name','id'])

```

```python
#scrapes the s at the end of all units

ss=[str(i).endswith('s') for i in ingredients_df.unit.values]
tt=[str(i)[:-1] for i in ingredients_df.unit[ss]]
ingredients_df.unit[ss]=tt

#converts the stupid units
#oz to g
ingredients_df.loc[ingredients_df['unit']=='oz','amount_metrics'] = (ingredients_df['amount']*28.3495) 
ingredients_df.loc[ingredients_df['unit']=='oz','unit_metrics'] = 'g' 

#lb to g
ingredients_df.loc[ingredients_df['unit']=='lb','amount_metrics'] = (ingredients_df['amount']*453.592) 
ingredients_df.loc[ingredients_df['unit']=='lb','unit_metrics'] = 'g' 

#cups to ml
ingredients_df.loc[ingredients_df['unit']=='c','amount_metrics'] = (ingredients_df['amount']*236.588) 
ingredients_df.loc[ingredients_df['unit']=='c','unit_metrics'] = 'mL' 

#tablespoon to mL (assuming US tablespoon) 17.7582
ingredients_df.loc[(ingredients_df['unit']=='T' )| (ingredients_df['unit']=='Tbsp'),'amount_metrics'] = (ingredients_df['amount']*17.7582) 
ingredients_df.loc[ingredients_df['unit']=='T','unit_metrics'] = 'mL' 

#teaspoon to mL (assuming US teaspoon)
ingredients_df.loc[ingredients_df['unit']=='t','amount_metrics'] = (ingredients_df['amount']*5.91939) 
ingredients_df.loc[ingredients_df['unit']=='t','unit_metrics'] = 'mL' 

#fluid ounce to mL
ingredients_df.loc[ingredients_df['unit']=='fluid ounce','amount_metrics'] = (ingredients_df['amount']*28.4131) 
ingredients_df.loc[ingredients_df['unit']=='fluid ounce','unit_metrics'] = 'mL' 

#blank to unit
ingredients_df.loc[ingredients_df['unit']=='','amount_metrics'] = ingredients_df['amount']
ingredients_df.loc[ingredients_df['unit']=='','unit_metrics'] = 'unit' 

```

```python
#replace unknown units with the original

ss=pd.isnull(ingredients_df.amount_metrics).values
ingredients_df.amount_metrics[ss]=ingredients_df.amount[ss]
ingredients_df.unit_metrics[ss]=ingredients_df.unit[ss]
```

```python
#print and save the shopping list for the week
file=open('ingredients_list.csv','w')

file.write('"amount","unit","ingredient"\n')

for name, group in ingredients_df.groupby(['id']):#loops through group by object by name of the group / and group
        #group.groupby(['unit_metrics'])
        ShortName= min(group.name.values, key=len)  
        for name2,group2 in group.groupby(['unit_metrics']):
            print int(np.float(group2.sum().amount_metrics)), name2, ShortName
            s=str(int(np.float(group2.sum().amount_metrics)))+','+ str(name2)+',' +str(ShortName)+'\n'
            file.write(s)
            
file.close()
```
